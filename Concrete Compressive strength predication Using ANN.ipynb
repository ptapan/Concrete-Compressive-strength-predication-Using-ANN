```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
```

```python
X_train = scalar.fit_transform(X_train)
X_test = scalar.transform(X_test)
```

## Using Neural Network

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow import keras
from tensorflow.keras.layers import Dense, Dropout
from keras import layers
from keras import Sequential
from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.wrappers.scikit_learn import KerasRegressor
```

# Hyperaparameter Tuing Using RandomSearch CV

# For batch Size and no of epochs

```python
 def create_model():
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1)
         ])
     
    # compile model
        model.compile(loss='mse', metrics=['mse'])
        return model
    
    
seed = 7
tf.random.set_seed(seed)   
model = KerasRegressor(build_fn=create_model)
# define the grid search parameters
batch_size = [8,16, 32, 64, 128, 256]
epochs = [180,200,250,300]
#optimizer = ['sgd', 'rmsprop', 'adam']
#earning_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
parameter = dict(batch_size=batch_size, epochs=epochs)
rnd_cv1 = RandomizedSearchCV(estimator=model, param_distributions=parameter, n_jobs=-1, cv=3)
rnd_cv1.fit(X_train, y_train)
rnd_cv1.best_params_
```

# Tuing optimization Learnig_rate

```python
 def create_model(learning_rate=3e-3):
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1)
         ])
        optimizer = keras.optimizers.Adam(lr=learning_rate)
    # compile model
        model.compile(loss='mse', metrics=['mse'],optimizer=optimizer)
        return model
    
    
seed = 7
tf.random.set_seed(seed)   
model = KerasRegressor(build_fn=create_model)
# define the grid search parameters
learning_rate = [0.001, 0.01, 0.1, 0.2, 0.3]
parameter = dict(learning_rate=learning_rate)
rnd_cv1 = RandomizedSearchCV(estimator=model, param_distributions=parameter, n_jobs=-1, cv=3)
rnd_cv1.fit(X_train, y_train,epochs=200,batch_size=64)
rnd_cv1.best_params_
```

## Building Model With Tuned Parameter

```python
def create_model(learning_rate=1e-1):
        
    # define model
        model = keras.Sequential([
        keras.Input(shape=X_train.shape[1]),
        Dense(64, activation="relu"),
        Dense(32, activation="relu"),
        Dense(16, activation="relu"),
        Dense(1)
         ])
        optimizer = keras.optimizers.Adam(lr=learning_rate)
    # compile model
        model.compile(loss='mse', metrics=['mse'],optimizer=optimizer)
        return model
    
    

tf.random.set_seed(seed)   
model = KerasRegressor(build_fn=create_model)

model.fit(X_train,y_train,epochs=200,batch_size=64)
```

```python
prediction = model.predict(X_test)
```

# MSE 


```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(y_test,prediction)
```
